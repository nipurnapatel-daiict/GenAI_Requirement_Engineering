```python
## invest framework

## Pipeline 3 - Take input={user_stories} from previous output

from langchain.prompts import ChatPromptTemplate

template = f"""
User Stories: {user_stories.content}

Introduction: Assume you are an AI-powered Requirements Analyst working for Retinodes Software Company. You are an expert in evaluating user stories
                  against the INVEST framework. Your goal is to assess the provided user stories and determine if they adhere to the INVEST principles.

Task: Evaluate each user story provided above against the INVEST framework.
    INVEST Framework Rules:
      Independent: All the stories that have been broken down are mostly independent, hence making sure that they can be parallel developed.
      Negotiable: Not every minute detail has been included in the cards, encouraging developers to seek further clarification and ensuring negotiability.
      Valuable: Stories have been written in regular business language that is equally valuable to both stakeholders and developers.
      Estimable: It is assumed that all the user stories mentioned below can be converted into working pieces of code within the working period of 1 to 3 weeks and hence are estimable and small.
      Small: All user stories should be considered small enough to fit into a 1-3 week iteration. **Assume that all user stories are small** (i.e., ready to be developed in a single iteration).
      Testable: It is also written in a way where the completion of the story and its working is quantifiable and can be tested, hence being testable.

    Contextual Information: NYU is an insurance company that aims to provide a comprehensive digital solution that enhances the agent-insurer relationship, enables customers to create
    custom insurance packages, and offers competitive pricing against other market players. The system should analyze existing insurance packages, understand market demands, and comply with legal and regulatory guidelines.
    User stories are being developed for the NYU Market-Driven system, an insurance platform.  The INVEST framework is used to ensure high-quality, actionable user stories.

Instructions:
  - For each of the at least 10 user stories provided, assess whether it meets each of the INVEST criteria (Independent, Negotiable, Valuable, Estimable, Testable).
  - Small: Assume all user stories are *Small* (i.e., they fit into a 1-3 week iteration and can be developed within that time).
  - Please provide a detailed evaluation for *each individual story*, following this *exact* format for every user story:

    User Story: [The complete user story text]

    Independent: [Yes/No/Partially]\n - [Explanation of why or why not]\n

    Negotiable: [Yes/No/Partially]\n - [Explanation of why or why not]\n

    Valuable: [Yes/No/Partially]\n- [Explanation of why or why not]\n

    Estimable: [Yes/No/Partially]\n - [Explanation of why or why not]\n

    Small: [Yes]\n - [Explanation: Assume all user stories are small and can be completed in a 1-3 week iteration.]\n

    Testable: [Yes/No/Partially]\n - [Explanation of why or why not]\n

Closing: After evaluating each story, provide an overall assessment of the quality of the user stories with respect to the INVEST framework. Identify any common issues or areas for improvement.
          Consider how the user stories could be improved to better adhere to the INVEST principles.

Please make sure to evaluate **all** the user stories and ensure the output is clear and comprehensive.
"""

prompt = ChatPromptTemplate.from_template(template)
prompt_template = prompt.format(user_stories={"content": user_stories})

example = f"""
AI:
{prompt_template}

HUMAN:
Validate INVEST criteria for each user story and provide a detailed evaluation with justification for each principle.
I need at least **10** user stories evaluated. Make sure to address each principle clearly and provide the necessary explanations. In particular, assume that all user stories are **Small** and do not need to be evaluated for size.
AI:
"""

validated_user_stories = llm.invoke(example)
print(validated_user_stories.content)
```

```python
## Extract user stories

## Pipeline 4 - Take input={validated_user_stories} from previous output

import re

template = f"""
Validated User Stories: {validated_user_stories.content}

 Introduction: Act as a data analyst. You are an expert in extracting user stories from a structured text.

 Task: Extract the user stories that have been validated as adhering to the INVEST principles.

 Contextual Information: The provided text contains user story evaluations against the INVEST framework. You need to identify and extract only those user stories that were assessed as meeting *all* the INVEST criteria (Independent, Negotiable, Valuable, Estimable, Testable).

 Instructions: Examine the provided text. For each user story that has *Yes* for *all* INVEST criteria (except Small, which is covered by Estimable), extract the *complete* user story text. Return the extracted user stories in a clear and readable format, number wise, separated by newlines or other delimiters.

 Closing: Provide the extracted user stories. If no user stories meet all the INVEST criteria, indicate that.
"""

prompt = ChatPromptTemplate.from_template(template)
extracted_data = llm.invoke(template)
print(extracted_data.content)


validated_stories = []

for story_block in extracted_data.content.split("---"):
    if not story_block.strip():
        continue

    match = re.search(r"User Story: (.+)\nIndependent: (Yes|No|Partially)\nNegotiable: (Yes|No|Partially)\nValuable: (Yes|No|Partially)\nEstimable: (Yes|No|Partially)\nSmall: (Yes|No|Partially)\nTestable: (Yes|No|Partially)", story_block, re.DOTALL)
    if match:
        user_story = match.group(1).strip()
        independent = match.group(2) == "Yes"
        negotiable = match.group(3) == "Yes"
        valuable = match.group(4) == "Yes"
        estimable = match.group(5) == "Yes"
        testable = match.group(6) == "Yes"

        if independent and negotiable and valuable and estimable and testable:
            validated_stories.append(user_story)

final_validated_output = "\n\n---\n\n".join(validated_stories)

print("Validated User Stories:\n")
print(final_validated_output)
```
